```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
landcovers1 = ["Alfalfa", "Other Pasture", "Lettuce", "Wheat", "Onions", "Other Truck Crops", "Corn", "Other Field Crops",
             "Subtropical", "Other"]
nClasses = 14
landcovers2 = ["Rice", "Safflower", "Corn", "Other Field crops", "Alfalfa", "Other Pasture", "Cucurbits", "Tomatoes", 
             "Other Truck Crops", "Almonds and pistachios", "Other Deciduous", "Subtropical", "Vineyards", "Other"]
```

```python
def calculate_f1(cf):
    weights = np.sum(cf, axis = 1)/np.sum(cf)
    recall = np.diag(cf) / np.sum(cf, axis = 1)
    precision = np.diag(cf) / np.sum(cf, axis = 0)
    f1 = (2*precision*recall)/(precision+recall)
    weighted_f1 = np.sum(weights*f1)
    return weighted_f1
```

```python
a = ['rf', '1d', '2d', 'paper', '3d1d']
```

```python
imperial_rf = np.load('imperial_rf.npy')
imperial_1d = np.load('imperial_1d.npy')
imperial_2d = np.load('imperial_2d.npy')
imperial_paper = np.load('imperial_paper.npy')
imperial_3d = np.load('imperial_3d1d.npy')
imperial_cfs = [imperial_rf, imperial_1d, imperial_2d, imperial_paper, imperial_3d]
```

```python
Yolo_rf = np.load('Yolo_rf.npy')
Yolo_1d = np.load('Yolo_1d.npy')
Yolo_2d = np.load('Yolo_2d.npy')
Yolo_paper = np.load('Yolo_paper.npy')
Yolo_3d = np.load('Yolo_3d1d.npy')
Yolo_cfs = [Yolo_rf, Yolo_1d, Yolo_2d, Yolo_paper, Yolo_3d]
```

```python
for i in range(5):
    acc = np.trace(imperial_cfs[i])/np.sum(imperial_cfs[i])
    f1 = calculate_f1(imperial_cfs[i])
    print(a[i],':',acc*100, f1*100)
    
    s = np.sum(imperial_cfs[i], axis = 1)
    cf1 = pd.DataFrame(imperial_cfs[i]/(s.reshape(10, 1))*100, columns=landcovers1, index = landcovers1)
    plt.figure(figsize = (35,15))
    sns.set(font_scale=2) 
    sns.heatmap(cf1, annot=True, annot_kws={"size": 25},fmt = '.0f')
    plt.savefig("imperial/imperial"+a[i]+".png")
```

```python
for i in range(5):
    acc = np.trace(Yolo_cfs[i])/np.sum(Yolo_cfs[i])
    f1 = calculate_f1(Yolo_cfs[i])
    print(a[i],':',acc*100, f1*100)
    s = np.sum(Yolo_cfs[i], axis = 1)
    cf1 = pd.DataFrame(Yolo_cfs[i]/(s.reshape(14, 1))*100, columns=landcovers2, index = landcovers2)
    plt.figure(figsize = (35,15))
    sns.set(font_scale=2) 
    sns.heatmap(cf1, annot=True, annot_kws={"size": 25},fmt = '.0f')
    plt.savefig("Yolo/Yolo"+a[i]+".png")
```
